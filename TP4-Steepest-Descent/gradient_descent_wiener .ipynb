```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import inv
from IPython.display import Image
```

## Implementar el filtrado óptimo Wiener con descenso por gradiente en la plantilla steepest_descent.py. Prestar atención al formato de salida:

```python
def steepest_descent(R, p, w0, mu, N):
    """Implementa el filtrado óptimo Wiener con descenso por gradiente.

    Argumentos:
        R: matriz de autocorrelación
        p: matriz de correlación cruzada
        w0: valor inicial de los coeficientes del filtro
        mu: tamaño de paso
        N: número máximo de iteraciones

    Devuelve:
        Una matriz de tipo np.array en cuyas filas están
        los coeficientes w para cada paso.
    """
    Wt = np.zeros((N, len(w0)))
    w_n = w0
    for i in range(N):
        # w(n+1) = w(n) - 0.5*mu*delta_J(w)
        delta_J = p-np.dot(R,w_n)
        w_aux = w_n + mu*delta_J
        Wt[i] = w_aux
        w_n = w_aux
    return Wt
```

## Probar la función con los argumentos dados:

```python
R = np.array([[2, 1],
             [1, 2]])
p = np.array([6,4])
w0 = np.array([0,0])
mu = 0.1
N = 1000
Wt = steepest_descent(R, p, w0, mu, N)
```

```python
w_1 = Wt[:,0]
w_2 = Wt[:,1]
```

### Representar los coeficientes w1 y w2 en función del tiempo:

#### Representar como referencia la solución del filtro óptimo Wiener correspondiente.

```python
def wiener_coefs(R, p):
    return np.dot(inv(R), p)
```

```python
w_o = wiener_coefs(R, p)
print(w_o)
```

## Representar la curva de error J(w):

```python
def get_J(w, var, p, R):
    J_w = []
    J_w.append(var)
    for i in range(len(w)):
        w_aux = w[i]
        w_dot_p = np.dot(p,w_aux)
        w_dot_R = np.dot(w_aux,R)
        J_aux = var-2*w_dot_p+np.dot(w_dot_R,w_aux)
        J_w.append(J_aux)
    return J_w
```

```python
var = 20
J_s = get_J(Wt, var, p, R)
```

```python
plt.figure(num=None, figsize=(12,6), dpi=80, facecolor='w', edgecolor='k')
plt.ylabel('J')
plt.xlabel('n')
plt.title('J(n)')
plt.grid('on')
plt.plot(J_s)
plt.show()
```

## Determinar los autovalores de R, y a partir de ellos determinar µmax. Para un valor de µ cercano a µmax representar la evolución de los coeficientes w(t):

```python
ev, _ = np.linalg.eig(R)
lambda_max = np.amax(ev)
mu_max = 2/lambda_max
print('mu_max: ', mu_max)
```

## Considerar el escenario de R y p variables en el tiempo. Explicar los beneficios del descenso por gradiente en comparación con el método de Wiener:

En el caso de tener $ R $ y $ \rho $ variables en el tiempo, se puede utilizar descenso por gradiente para ajustar los coeficientes del filtro en tiempo real. Ya que en tiempo real pueden aparecer perturbaciones que modifiquen los $ w_{o} $ y de esta manera se pueden ajustar los coeficientes en tiempo real de manera iterativa. Si se utiliza el metodo de Wiener, se tendria que correr el algoritmo cada vez que se tengan que ajustar nuevamente los valores. Ademas, el metodo de Wiener es mas complejo computacionalemnte por lo que si se quiere llegar a una solucion aceptable rapidamente, el metodo de descenso por gradietne es una mejor eleccion. Una posible deventaja de este metodo es que si los $ R $ y $ \rho $ varian rapidamente en el tiempo, es posible que los coeficientes del filtro no converjan o demore mas de lo permitido.
